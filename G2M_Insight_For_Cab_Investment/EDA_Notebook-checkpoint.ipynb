```python
# combining all datasets together based on foreign keys
master_data = (df_transaction_id.merge(df_cab_data, how='inner')  # Merging transaction and cab data
                .merge(df_customer_id, how='inner')  # Merging with customer data
                .merge(df_city, how='inner'))  # Merging with city data

# Add a 'Net Profit' column based on 'Price Charged' and 'Cost of Trip
master_data.insert(loc=9, column='Net Profit', value=master_data['Price Charged'] - master_data['Cost of Trip'])

# Calculate and add a column for the 'Rate per KM'
master_data.insert(loc=6, column='Price per KM', value=master_data['Price Charged'] / master_data['KM Travelled'])

# Calculate and add a 'Net Profit per KM' column
master_data.insert(loc=7, column='Net Profit per KM', value=master_data['Net Profit'] / master_data['KM Travelled'])

# Initialize index variable to iterate over rows
index = 0
for date in master_data['Date of Travel']:
    days_since_2016 = date - 42364  # Calculating days since start of 2016
    if days_since_2016 >= 0 and days_since_2016 <= 335:
        master_data.at[index, 'Year'] = 2016  # Assigning year 2016
    elif days_since_2016 <= 700:
        master_data.at[index, 'Year'] = 2017  # Assigning year 2017
    else:
        master_data.at[index, 'Year'] = 2018  # Assigning year 2018
    index = index + 1

# Count the number of rides for each 'Customer ID' and reset index to make it a DataFrame
rides_summary = master_data['Customer ID'].value_counts().rename_axis('Customer ID').reset_index(name='Number of Rides')

# Merge the ride count data back into the main DataFrame
master_data = pd.merge(master_data, rides_summary, on='Customer ID')

master_data
```

```python

# THE FIRST HYPOTHESIS



# Hypothesis: Yellow Cab is expected to have higher Net profit margins compared to Pink Cab based on the average Net profit per transaction over a three-year period.

# The objective is to determine whether Yellow Cab consistently generates higher Net profit margins than Pink Cab, providing insights into their financial performance and potential investment viability.

# Analyzing Net profit data for Pink Cab
filtered_data = df_master_data.loc[(df_master_data['Company'] == 'Pink Cab') & 
                                (df_master_data['Year'] == 2016)]
pink_2016 = filtered_data['Net Profit'].mean()
print("Pink Cab average Net profit per transaction in 2016:", pink_2016)

filtered_data = df_master_data.loc[(df_master_data['Company'] == 'Pink Cab') & 
                                (df_master_data['Year'] == 2017)]
pink_2017 = filtered_data['Net Profit'].mean()
print("Pink Cab average Net profit per transaction in 2017:", pink_2017)

filtered_data = df_master_data.loc[(df_master_data['Company'] == 'Pink Cab') & 
                                (df_master_data['Year'] == 2018)]
pink_2018 = filtered_data['Net Profit'].mean()
print("Pink Cab average Net profit per transaction in 2018:", pink_2018)
print('\n')

# Analyzing Net profit data for Yellow Cab
filtered_data = df_master_data.loc[(df_master_data['Company'] == 'Yellow Cab') & 
                                (master_data['Year'] == 2016)]
yellow_2016 = filtered_data['Net Profit'].mean()
print("Yellow Cab average Net profit per transaction in 2016:", yellow_2016)

filtered_data = df_master_data.loc[(df_master_data['Company'] == 'Yellow Cab') & 
                                (df_master_data['Year'] == 2017)]
yellow_2017 = filtered_data['Net Profit'].mean()
print("Yellow Cab average Net profit per transaction in 2017:", yellow_2017)

filtered_data = df_master_data.loc[(df_master_data['Company'] == 'Yellow Cab') & 
                                (df_master_data['Year'] == 2018)]
yellow_2018 = filtered_data['Net Profit'].mean()
print("Yellow Cab average Net profit per transaction in 2018:", yellow_2018)

# Creating a data frame for Net profit comparison between Pink Cab and Yellow Cab
Net_profit_data = pd.DataFrame({
    'Year': [2016, 2017, 2018, 2016, 2017, 2018],
    'Company': ['Pink Cab', 'Pink Cab', 'Pink Cab', 'Yellow Cab', 'Yellow Cab', 'Yellow Cab'],
    'Net Profit': [pink_2016, pink_2017, pink_2018, yellow_2016, yellow_2017, yellow_2018]
})

# Creating a pivot table for better visualization of Net profit over the years
pivot_table = pd.pivot_table(Net_profit_data, values='Net Profit', index='Year', columns='Company')

# Plotting the Net profit comparison as a line graph
pivot_table.plot(kind='line',color=['pink','yellow'])
plt.xlabel('Year')
plt.ylabel('Average Net Profit per Transaction')
plt.title('Average Net Profit per Transaction for Pink Cab and Yellow Cab (2016-2018)')
plt.xticks(rotation=90)
plt.show()

# Visualizing the Net profit distribution for Pink Cab and Yellow Cab
pink_data = df_master_data[df_master_data['Company'] == 'Pink Cab']
yellow_data = df_master_data[df_master_data['Company'] == 'Yellow Cab']

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 5))


# Histogram for Yellow Cab
ax2.hist(yellow_data['Net Profit'], bins=10, color='yellow')
ax2.set_title('Net Profit Distribution for Yellow Cab')
ax2.set_xlabel('Net Profit')
ax2.set_ylabel('Frequency')

# Histogram for Pink Cab
ax1.hist(pink_data['Net Profit'], bins=10, color='pink')
ax1.set_title('Net Profit Distribution for Pink Cab')
ax1.set_xlabel('Net Profit')
ax1.set_ylabel('Frequency')

plt.show()

# Concl :  

# The analysis indicates that the most frequent Net profit margins for Pink Cab per transaction are centered around zero, suggesting limited profitability.
# In contrast, Yellow Cab's Net profit margins tend to be greater than zero, indicating stronger and more consistent profitability.
# This supports the hypothesis that Yellow Cab has higher Net profit margins, making it a potentially better investment opportunity.
```

```python
# THE FIFTH HYPOTHESIS 


# Hypothesis: Yellow Cab is expected to have a more balanced profit distribution across its major locations.
# A diverse profit distribution across different cities may provide stability and may help the company intimes of financial crisis.


# Creating side-by-side histograms to visualize city frequency distribution for both cab companies
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Loop through each unique cab company to generate frequency histograms
for i, company in enumerate(df_master_data['Company'].unique()):
    city_counts = df_master_data.loc[df_master_data['Company'] == company, 'City'].value_counts()
    axs[i].hist(city_counts.index, weights=city_counts.values, bins=len(city_counts), color='green',edgecolor='black')
    axs[i].set_title(f'City Frequencies for {company}')
    axs[i].set_xlabel('City')
    axs[i].set_ylabel('Frequency')
    axs[i].tick_params(axis='x', rotation=90)

# Adjust layout for clear visualization
plt.tight_layout()
plt.show()

# Calculating total profit from Los Angeles for Pink Cab
pink_LA = df_master_data.loc[(df_master_data['Company'] == 'Pink Cab') 
                                        & (df_master_data['City'] == 'LOS ANGELES CA'), 'Net Profit'].sum()

# Calculating total profit from New York City for Pink Cab
pink_NYC = df_master_data.loc[(df_master_data['Company'] == 'Pink Cab') 
                                        & (df_master_data['City'] == 'NEW YORK NY'), 'Net Profit'].sum()

# Calculating total profit from San Diego for Pink Cab
pink_SD = df_master_data.loc[(df_master_data['Company'] == 'Pink Cab') 
                                        & (df_master_data['City'] == 'SAN DIEGO CA'), 'Net Profit'].sum()

# Plotting profit distribution for Pink Cab's top three cities
pink_data = [pink_LA, pink_NYC, pink_SD]
pink_labels = ['Los Angeles, CA', 'New York City, NY', 'San Diego, CA']
plt.figure()
plt.pie(pink_data, labels=pink_labels, autopct='%1.1f%%', startangle=90,colors=['green','blue','orange'])
plt.title('Distribution of Profits per City for Pink Cab')
plt.axis('equal')  # Ensures pie chart is circular

plt.show()

# Calculating total profit from New York City for Yellow Cab
yellow_NYC = df_master_data.loc[(df_master_data['Company'] == 'Yellow Cab') 
                                        & (df_master_data['City'] == 'NEW YORK NY'), 'Net Profit'].sum()

# Calculating total profit from Chicago for Yellow Cab
yellow_CHI = df_master_data.loc[(df_master_data['Company'] == 'Yellow Cab') 
                                        & (df_master_data['City'] == 'CHICAGO IL'), 'Net Profit'].sum()

# Calculating total profit from Washington D.C. for Yellow Cab
yellow_DC = df_master_data.loc[(df_master_data['Company'] == 'Yellow Cab') 
                                        & (df_master_data['City'] == 'WASHINGTON DC'), 'Net Profit'].sum()

# Plotting profit distribution for Yellow Cab's top three cities
yellow_data = [yellow_NYC, yellow_CHI, yellow_DC]
yellow_labels = ['New York City, NY', 'Chicago, IL', 'Washington D.C.']
plt.figure()
plt.pie(yellow_data, labels=yellow_labels, autopct='%1.1f%%', startangle=90,colors=['green','blue','orange'])
plt.title('Distribution of Profits per City for Yellow Cab')
plt.axis('equal')  # Ensures pie chart is circular

plt.show()

# Final Observations: Yellow Cab has a high concentration of profits from New York City, contributing about 80%.
# Final Observations: Contrary to the hypothesis, Pink Cab shows a more balanced profit distribution across its top cities.
# Conclusion: The analysis suggests Pink Cab may be better positioned to withstand economic challenges in specific cities.
```
